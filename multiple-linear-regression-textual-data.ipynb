```python
import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/car-price-prediction-mlr-textual-data/Used Car Dataset.csv")
df.head()
```

* In above DataFrame  car_name,insurance_validity input columns required nominal encoding
* ownsership,transmission,fuel_type required ordered encoding

# Ordinal Encoding

```python
df["fuel_type"].value_counts()

#see total number of distict values present in fuel_type feature
```

```python
df["ownership"].value_counts()

#see total number of distict values present in ownership input column
```

```python
df["transmission"].value_counts()

#see total no of distinct values present in the transmission
```

* CNG > Diesel > Petrol
* First Owner > Second Owner > Third Owner
* Automatic > Manual

```python
df2=pd.DataFrame()
df2=pd.concat([df2,df["fuel_type"],df["ownership"],df["transmission"]],axis=1)
print(df2)

#concate 3rd ,6th ,7th column together and make different DataFrame 

#concatination syntex:-
#pandas.concat([target_df,column 1,column 2,...,column n] , axis=0/1)
#axis=0 means concatination on x-axis
#axis=1 means concatination on y-axis
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[["Petrol","Diesel","CNG"],["Third Owner","Second Owner","First Owner"],["Manual","Automatic"]])

#making object or Ordinal Encoding Class
```

```python
fot_array=oe.fit_transform(df2)
fot_array

#apply ordinal encoding
#it returns array
```

```python
fot_df=pd.DataFrame(fot_array,columns=["fuel_type","ownership","transmission"])
fot_df

#convert above array into dataframe
```

# Nominal Encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop="first")

#making object of OneHotEncodig clas
```

```python
df["insurance_validity"].head()
```

```python
df.iloc[:,2:3].value_counts()

#see how many distinct values are present in insurance_validity
```

```python
in_v_sparce_arr = ohe.fit_transform(df.iloc[:,2:3])
print(in_v_sparce_arr)
in_v_arr=in_v_sparce_arr.toarray()
print(in_v_arr)

#do OneHotEncoding
#it give sparce matrics stored in in_v_sparce_arr
# in_v_sparce_arr means insurance_validity_sparce_array
#convert sparce matrics to an array using .toarray() function and stored in in_v_arr array
#in_v_arr means insurance_validity_array
```

```python
in_v_arr_df=pd.DataFrame(in_v_arr,columns=["insurance_validity_Third Party insurance","insurance_validity_Zero Dep","insurance_validity_Third Party"])
in_v_arr_df.head()

#convert above in_v_arr to DataFrame named in_v_arr_df
#in_v_arr_df means insurance_validity_array_dataframe
```

# Ordinal Encoding With Top Categories (car_name)

```python
car_count=df["car_name"].value_counts()
car_count

# see which different values are present in car_name column
```

make all the cars having <= 5 value, put them together and named as others

# Using Pandas get_dummies() function

```python
thresould=5
replacable=car_count[car_count <= thresould]
replacable

#fatch all the columns which are replacable
```

```python
replacable_car_array = replacable.index
print(replacable_car_array)

#fatch all the name of the car form replacable dataframe using .index
```

```python
replaced_df = df["car_name"].replace(replacable_car_array,"others")
replaced_df.head()

#replace replacable car_name with others
#replace not modifies actual data
```

```python
replaced_df.value_counts()

#now again see what different values are present in the car_name column of replaced_df datafram
```

```python
new_car_name_df=pd.get_dummies(replaced_df)
print(new_car_name_df.head())
print(new_car_name_df.shape)

#do nominal encoding using get_dummies function of pandas
```

# Concatination of all the transformed columns

```python
df.head()
```

```python
new_df=pd.DataFrame()
new_df=pd.concat([new_df,new_car_name_df,df["registration_year"],in_v_arr_df,fot_df,df["seats"],df["kms_driven"],df["manufacturing_year"],df["mileage(kmpl)"],df["engine(cc)"],df["max_power(bhp)"],df["torque(Nm)"],df["price(in lakhs)"]],axis=1)
print(new_df.head())

#concate all the columns which we transformed and which remain as it is in new datafram
```

```python
print("shape of original dataset: ",df.shape)
print("shape of modefied dataset: ",new_df.shape)
```

# Model Training Part

```python
new_df.iloc[:1,:]
```

```python
x=new_df.iloc[:,:-1]
x.head()

#extract x from dataframe
```

```python
y=new_df.iloc[:,-1]
y

#extract y from datafram
```

```python
x=x.to_numpy()
print(x)

#convert x into numpy array
```

```python
y=y.to_numpy()
print(y)

#convert y into the numpy array
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=22)

#aplitting training data and testing data from x and y
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

#make the object of linear regression class
```

```python
lr.fit(x_train,y_train)

#model traininig
```

```python
y_predict=lr.predict(x_test)
print(y_predict)

#model testing
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)

#model evaluation
```
