# n-1 One-Hot-Encoding

* here n is the no of distinct value in a column.
* as of now remember to remove **Multicolinearity problem** we use this methode
* in this methode we only create n-1 columns using get_dummies() function for n different values
* so that, the sum is always 1(as we discussed above) can be solve,

**How it can work**

* fuel_CNG___fuel_Diesel___fuel_LPG___fuel_Petrol
* False______True__________False______True

* if car is Petrol then,
* False______False_________False______False

* above example it make the column only for 3 distict values suppose it will be fuel_CNG, fuel_Diesel, fuel_LPG
* table can be,

1. if the car is Diesel then,
* fuel_CNG___fuel_Diesel___fuel_LPG
* False______True__________False

2. if car is Petrol then,
* False______False_________False

* so when value of fuel_Perol is required then,
**for case 1 it is False and for case 2 it is True**
* if True is found in the row for the fuel_xyz column then predicted value is False
* if True is **not** found in the row for the fuel_xyz column then predicted value is True

# One-Hot_Encoding using sci-kit Learn

* **Theory**

> * We used sci-kit learn OrdinalEncoder class for encode the data
> * With this class we only have one issue,
> * For doing encoding with this class we have to follow below steps,
>> 1. separate columns from DataFrame such a way that columns belongs to ordinal encoding is in separate group, columns belongs to nominal encoding / one hot encoding (covering soon) are in a different group.
>> 2. apply encoding algorithm.
>> 3. combine all the transformed column with untransformed column to make new DataFrame.

* **We follow above procedure for now but for avoiding this many steps we have one another concept named COLUMN TRANSFORMER (see later)**

**Step: 1 separate columns from DataFrame such a way that columns belongs to ordinal encoding is in separate group, columns belongs to nominal encoding / one hot encoding (covering soon) are in a different group.**

```python
df2=df.iloc[:,2:4]
df2.head()
```

**Step: 2 apply encoding algorithm.**


# Without droping

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
s_array=ohe.fit_transform(df2)

#it provieds sparce matrix
```

```python
print(s_array.toarray())
print()
print(s_array[:1,:].toarray())
print()
print(s_array[:1,:].toarray().shape)

#shape has to be 4+5=9
```

```python
ohe = OneHotEncoder(drop="first")

# for include droping only need to give perameter to OneHotEncoder that is drop="first"
```

```python
s_array=ohe.fit_transform(df2)
s_array

#it provieds sparce matrix
```
