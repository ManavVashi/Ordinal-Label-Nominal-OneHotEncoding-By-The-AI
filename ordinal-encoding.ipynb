> **if your data in a perticular column is ordered then we use ordinal encoding**

* What is the meaninig of ordered data??
> * see the above DataFrame,
> * in review column there are 3 values 
>> 1. Poor
>> 2. Good
>> 3. Average
> * Good > Average > Poor
> * in short orderd means importance of the values in a same column are different.
> * also PG > UG > School

* **We are not going to apply ordered encoding for column named gender as male and female have same priority.**

* **note: Our data set does not have noise/incosistant values so we directly use it for understanding ordinal encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])

#importing class of OrdinalEncoder defined by sci-kit learn
```

```python
#below line gives us error

oe.fit(df)
```

* **Explaination**
> * We used sci-kit learn OrdinalEncoder class for encode the data
> * With this class we only have one issue,
> * For doing encoding with this class we have to follow below steps,
>> 1. separate columns from DataFrame such a way that columns belongs to ordinal encoding is in separate group, columns belongs to nominal encoding / one hot encoding (covering soon) are in a different group.
>> 2. apply encoding algorithm.
>> 3. combine all the transformed column with untransformed column to make new DataFrame.

* **We follow above procedure for now but for avoiding this many steps we have one another concept named COLUMN TRANSFORMER (see later)**

**Step: 1 separate columns from DataFrame such a way that columns belongs to ordinal encoding is in separate group, columns belongs to nominal encoding / one hot encoding (covering soon) are in a different group.**

```python
df2=df.iloc[:,2:4]
df2.head()
```

**Step: 2 apply encoding algorithm.**

```python
#below line gives error as .fit() function take array not DataFrame

oe.fit(df2)
```

```python
arr_df2 = oe.transform(df2)
print(arr_df2[:10,])
```

* **Instead of doing fit and transform seperatly you can also use fit_transorm() methode directly** 

```python
arr_df2_test = oe.fit_transform(df2)
print(arr_df2_test [:10])
```

**You can see same output is generated**

# ***small twist in the story***

* When we talk about independent variables / input columns then ordinal encoding works for ordered datapoint values
* But if the column is dependent column / dependent variable / target column then we have to use label encoding.

**Note: a point to be noted that for input columns we only use ordinal/nominal encoding and for target column we only use label encoding**
